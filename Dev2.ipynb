```python
import mne
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("Step 1: Load the dataset from PhysioNet")
# Assuming we are using a motor imagery dataset from PhysioNet (link provided in your file)
eeg_data = mne.datasets.eegbci.load_data(subject=1, runs=[6, 10, 14])

print("Step 2: Load raw data")
raw = mne.io.read_raw_edf(eeg_data[0], preload=True)

print("Step 3: Visualize the raw data")
raw.plot(scalings='auto', title='Raw EEG Data')

print("Step 4: Apply bandpass filtering (1-40 Hz)")
raw_filtered = raw.filter(l_freq=1.0, h_freq=40.0)

print("Step 5: Visualize the filtered data")
raw_filtered.plot(scalings='auto', title='Filtered EEG Data')

print("Step 6: Feature Extraction (Optional)")
# Example: Compute signal power in different frequency bands
# We can use a Fourier Transform or Wavelet Transform to get signal power

# Plot the power spectral density (PSD)
# plt.figure()
# plt.semilogy(freqs, psds.mean(axis=0))
# plt.title('Power Spectral Density (PSD)')
# plt.xlabel('Frequency (Hz)')
# plt.ylabel('Power')
# plt.show()
```
